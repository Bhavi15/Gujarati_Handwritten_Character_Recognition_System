```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
from tensorflow.keras.optimizers import RMSprop
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
import cv2
import os
import keras
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPool2D , Flatten
from keras.applications.vgg16 import VGG16
```

```python
train = ImageDataGenerator(rescale = 1/255)
test = ImageDataGenerator(rescale = 1/255)
```

```python
img = r'C:\Users\Acer-PC\Desktop\Project3132465465\test\char_1\15.png'

image = cv2.imread(img)
print(image.shape)
```

```python
train_data= train.flow_from_directory('train/',
                                         target_size = (32,32),
                                         batch_size = 10,
                                         class_mode = 'categorical')
test_data = train.flow_from_directory('test/',
                                         target_size = (32,32),
                                         batch_size = 10,
                                         class_mode = 'categorical')
batchX, batchy = train_data.next()
print('Batch shape=%s, min=%.3f, max=%.3f' % (batchX.shape, batchX.min(), batchX.max()))
```

```python
train_data.class_indices
```

```python
base_model = VGG16(input_shape = (32, 32, 3), # Shape of our images
include_top = False, # Leave out the last fully connected layer
weights = 'imagenet')
```

```python
for layer in base_model.layers:
    layer.trainable = False
```

```python

x = keras.layers.Flatten()(base_model.output)

x = keras.layers.Dense(512, activation='relu')(x)
x = keras.layers.Dropout(0.5)(x)

x = keras.layers.Dense(1, activation='softmax')(x)

model = tf.keras.models.Model(base_model.input, x)

model.compile(optimizer = tf.keras.optimizers.RMSprop(learning_rate=0.0001), loss = 'categorical_crossentropy',metrics = ['acc'])
```

```python
vgghist = model.fit(train_data, validation_data = test_data, steps_per_epoch = 100 , epochs = 1)
```

```python
import time
time1 = time.time()
time2 = time.time()
score = model.evaluate(test_data, verbose=0)
time3 = time.time()
print("Train time = ", (time2 - time1)/60, "min", "\nTest time = ", (time3 - time2)/ 60, "min", "\n Test loss: ", score[0], " Test accuracy: ", score[1])
```
